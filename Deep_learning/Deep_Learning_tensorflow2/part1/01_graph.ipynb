# Graph

Working on the default graph

```python
import tensorflow as tf
from pprint import pprint
```

```python
#在原始預設的Graph上建立張量架構
a = tf.Variable(2.0, dtype = tf.float32)
b = tf.Variable(3.0, dtype = tf.float32, name = 'var_b')
c = tf.Variable(5.0, dtype = tf.float32, name = 'var_c')

init = tf.global_variables_initializer()
```

```python
#<變數型態 變數名稱 張量維數 數據類型>
pprint(tf.global_variables())
```

```python
tf.reset_default_graph()
```

```python
with tf.Session() as sess:
    
    sess.run(init)
    print(sess.run((a + b) * c))
```

Define your own graph


```python
g = tf.Graph() #定義一個名為g的Graph物件
```

```python
with g.as_default():  #在g這張設計圖上建立張量架構
    d = tf.Variable(1.0, dtype = tf.float32, name = 'd')
    e = tf.Variable(4.0, dtype = tf.float32, name = 'e')
    f = tf.Variable(7.0, dtype = tf.float32, name = 'f')
    
    pprint(tf.global_variables())
```

```python
pprint(tf.global_variables())
```

```python
with tf.Session(graph = g) as sess:
    init = tf.global_variables_initializer()
    sess.run(init)
    print(sess.run((d + e) * f))
```
