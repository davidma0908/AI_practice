```python
import tensorflow as tf
from pprint import pprint
```

```python
g = tf.Graph()

with g.as_default():
    
    X = tf.placeholder(tf.float32, [None, 784], name = 'input')
    y = tf.placeholder(tf.float32, [None, 10], name = 'label')
    

    h1 = tf.layers.dense(X, units = 25, activation = tf.nn.relu)
    h2 = tf.layers.dense(h1, units = 25, activation = tf.nn.relu, name = 'h2')
    h3 = tf.layers.dense(h2, units = 30, activation = tf.nn.relu, name = 'h3')
    h4 = tf.layers.dense(h3, units = 30, activation = tf.nn.relu, name = 'h4')


    output = tf.layers.dense(h4, units = 10)


    loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=output, labels=y))


    correct_prediction = tf.equal(tf.argmax(tf.nn.softmax(output),1), tf.argmax(y,1))
    compute_acc = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))


    train_step = tf.train.GradientDescentOptimizer(0.001).minimize(loss)
    
    pprint(tf.global_variables())
```

```python
g = tf.Graph()

with g.as_default():
    
    with tf.name_scope('placeholder'):
        X = tf.placeholder(tf.float32, [None, 784], name = 'input')
        y = tf.placeholder(tf.float32, [None, 10], name = 'label')
    
    with tf.variable_scope('hidden_block1'):
        h1 = tf.layers.dense(X, units = 25, activation = tf.nn.relu)
        h2 = tf.layers.dense(h1, units = 25, activation = tf.nn.relu, name = 'h2')
    
    with tf.variable_scope('hidden_block2'):
        h3 = tf.layers.dense(h2, units = 30, activation = tf.nn.relu, name = 'h3')
        h4 = tf.layers.dense(h3, units = 30, activation = tf.nn.relu, name = 'h4')

    
    with tf.variable_scope('output'):
        output = tf.layers.dense(h4, units = 10)
        
    with tf.name_scope('cross_entropy'):
        loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=output, labels=y))

    with tf.name_scope('accuracy'):
        correct_prediction = tf.equal(tf.argmax(tf.nn.softmax(output),1), tf.argmax(y,1))
        compute_acc = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))

    with tf.name_scope('train'):
        train_step = tf.train.GradientDescentOptimizer(0.001).minimize(loss)
    
    pprint(tf.global_variables())
```
